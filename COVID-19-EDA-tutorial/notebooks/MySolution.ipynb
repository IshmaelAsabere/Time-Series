## Access the data

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set style & figures inline
sns.set()
%matplotlib inline
```

```python
# Data urls
base_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/'
confirmed_cases_data_url = base_url + 'time_series_covid19_confirmed_global.csv'
death_cases_data_url = base_url + 'time_series_covid19_deaths_global.csv'
recovery_cases_data_url = base_url+ 'time_series_covid19_recovered_global.csv'

# Import datasets as pandas dataframes
raw_data_confirmed = pd.read_csv(confirmed_cases_data_url)
raw_data_deaths = pd.read_csv(death_cases_data_url)
raw_data_recovered = pd.read_csv(recovery_cases_data_url)
```

## Number of cases by country

### Confirmed Cases by country

```python
# Group by region (also drop 'Lat', 'Long' as it doesn't make sense to sum them here)
confirmed_country = raw_data_confirmed.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)
confirmed_country.head()
```

### Deaths by country

```python
# Group by region (also drop 'Lat', 'Long' as it doesn't make sense to sum them here)
deaths_country = raw_data_deaths.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)
deaths_country.head()
```

### Recoveries by country

```python
# Group by region (also drop 'Lat', 'Long' as it doesn't make sense to sum them here)
recovery_country = raw_data_recovered.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)
recovery_country.head()
```

## Visualise the Data

```python
# Transpose the data and index it by date
confirmed_country = confirmed_country.transpose()
confirmed_country.head()
```

```python
# We can use pandas to turn it into a DateTimeIndex:
# Set index as DateTimeIndex
datetime_index = pd.DatetimeIndex(confirmed_country.index)
confirmed_country.set_index(datetime_index, inplace=True)
# Check out index
confirmed_country.index
```

### Plot Top 5 countries Globally

```python
# Plot time series of several countries of interest
poi = ['China', 'US', 'Italy', 'France', 'Spain', 'Australia']

# Plot time series of several countries of interest
confirmed_country[poi].plot(figsize=(20,10), linewidth=2, marker='.', colormap='brg', fontsize=20)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Reported Confirmed cases count', fontsize=20);
plt.title('Reported Confirmed Cases Time Series', fontsize=20);
```

## Growth Curves

```python
# Split-Apply-Combine
#deaths_country = raw_data_deaths.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)
confirmed_country = raw_data_confirmed.groupby(['Country/Region']).sum().drop(['Lat', 'Long'], axis=1)

# Transpose
confirmed_country = confirmed_country.transpose()

# Set index as DateTimeIndex
datetime_index = pd.DatetimeIndex(confirmed_country.index)
confirmed_country.set_index(datetime_index, inplace=True)

# Check out head
confirmed_country.head()
```

```python
# Loop over columns & set values < 500 to None
for col in confirmed_country.columns:
    confirmed_country.loc[(confirmed_country[col] < 300),col] = None

# Check out tail
confirmed_country.tail()
```

### What happened after Confirmed Cases exceeded 300 in Top 5 Countries - Globally

```python
# Plot time series of several countries of interest
poi = ['US', 'Italy', 'France', 'Australia']
confirmed_country[poi].plot(figsize=(20,10), linewidth=2, marker='.', colormap='brg', fontsize=20)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Number of Confirmed Cases', fontsize=20);
plt.title('Time Series after Confirmed cases hit 300', fontsize=20);
```

### What happened after Confirmed Cases ≥ 300 in African Countries

```python
# Plot time series of several countries of interest
poi_afr = ['US', 'Italy', 'France', 'Australia']
confirmed_country[poi_afr].plot(figsize=(20,10), linewidth=2, marker='.', colormap='brg', fontsize=20)
plt.xlabel('Date', fontsize=20);
plt.ylabel('Number of Confirmed Cases', fontsize=20);
plt.title('Time Series after hitting 300 Confirmed Cases', fontsize=20);
```
